1) Find the proportion of main colors in a picture. As discussed in the class, this assignment will help you to practice handling images and pixels, numpy arrays, and unsupervised clustering

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
image = plt.imread('pic.jpg')
# image = image/255
# print(np.sum(image))
# print(image)
# reshape the image to a 2D array of pixels and 3 color values (RGB)
pixels = image.reshape(-1,3)
# print(pixels)

kmeans = KMeans(n_clusters=4).fit(pixels)
labels = kmeans.labels_
centers = kmeans.cluster_centers_
# centers =np.uint8(centers*255)
centers =np.uint8(centers)
print(centers)
segmented_image = centers[labels]
segmented_image=segmented_image.reshape(image.shape)
print(segmented_image)
labels_counted = np.bincount(labels)
total_pixels = np.sum(labels_counted)
print(total_pixels)

proportions = labels_counted / total_pixels

# Print proportions
for i, proportion in enumerate(proportions):
    print(f"Color {i} proportion: {proportion}")

plt.figure(figsize=(12, 6))
# bar_colors = centers / 255.0  # Normalize to [0, 1] range
bar_colors = centers / 255.0  # Normalize to [0, 1] range
plt.bar(range(4), proportions,color=bar_colors)
plt.xlabel('Cluster Index')
plt.ylabel('Proportion')
plt.title('Proportion of Main Colors')

for i, proportion in enumerate(proportions):
    plt.text(i, proportion, f"{proportion:.4f}", ha='center', va='bottom')

plt.show()
```
